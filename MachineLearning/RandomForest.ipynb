```python
# Loop through the 'station_line' column and assign a 1 or 0 to the new line column depending on if the station has the line
for i, row in yankees.iterrows():
    if '1' in row['station_line']:
        yankees.loc[i, 'Line_1'] = 1
    else:
        yankees.loc[i, 'Line_1'] = 0
    if '2' in row['station_line']:
        yankees.loc[i, 'Line_2'] = 1
    else:
        yankees.loc[i, 'Line_2'] = 0
    if '4' in row['station_line']:
        yankees.loc[i, 'Line_4'] = 1
    else:
        yankees.loc[i, 'Line_4'] = 0
    if '5' in row['station_line']:
        yankees.loc[i, 'Line_5'] = 1
    else:
        yankees.loc[i, 'Line_5'] = 0
    if '6' in row['station_line']:
        yankees.loc[i, 'Line_6'] = 1
    else:
        yankees.loc[i, 'Line_6'] = 0
    if 'A' in row['station_line']:
        yankees.loc[i, 'Line_A'] = 1
    else:
        yankees.loc[i, 'Line_A'] = 0
    if 'B' in row['station_line']:
        yankees.loc[i, 'Line_B'] = 1
    else:
        yankees.loc[i, 'Line_B'] = 0
    if 'C' in row['station_line']:
        yankees.loc[i, 'Line_C'] = 1
    else:
        yankees.loc[i, 'Line_C'] = 0
    if 'D' in row['station_line']:
        yankees.loc[i, 'Line_D'] = 1
    else:
        yankees.loc[i, 'Line_D'] = 0
    if 'G' in row['station_line']:
        yankees.loc[i, 'Line_G'] = 1
    else:
        yankees.loc[i, 'Line_G'] = 0
    if 'J' in row['station_line']:
        yankees.loc[i, 'Line_J'] = 1
    else:
        yankees.loc[i, 'Line_J'] = 0
    if 'L' in row['station_line']:
        yankees.loc[i, 'Line_L'] = 1
    else:
        yankees.loc[i, 'Line_L'] = 0
    if 'Z' in row['station_line']:
        yankees.loc[i, 'Line_Z'] = 1
    else:
        yankees.loc[i, 'Line_Z'] = 0
```

```python
# Dropping unneeded columns
yankees = yankees.drop(columns=['CMPLNT_NUM','RPT_DT', 'KY_CD', 'PD_CD', 'PD_DESC','DATE', 'ADDR_PCT_CD', 'LOC_OF_OCCUR_DESC', 'PREM_TYP_DESC', 'JURIS_DESC', 'SUSP_AGE_GROUP', 'SUSP_RACE','SUSP_SEX', 'PATROL_BORO', 'VIC_AGE_GROUP', 'VIC_RACE','VIC_SEX', 'STATION_NAME', 'station_line'])
```

```python
# Breaking out the column into its individual datetime components
yankees['year'] = yankees['DATE_AND_TIME'].dt.year
yankees['month'] = yankees['DATE_AND_TIME'].dt.month
yankees['day']= yankees['DATE_AND_TIME'].dt.day
yankees['hour'] = yankees['DATE_AND_TIME'].dt.hour
yankees['minute'] = yankees['DATE_AND_TIME'].dt.minute
yankees['second'] = yankees['DATE_AND_TIME'].dt.second
```

```python
# Create a new column for the number of days in the month
import numpy as np
yankees['days_in_month'] = np.where(yankees['month'] == 2, 28, np.where(yankees['month'] == 4, 30, np.where(yankees['month'] == 6, 30, np.where(yankees['month'] == 9, 30, np.where(yankees['month'] == 11, 30, 31)))))
```

```python
#create cyclical encodings
yankees['year_sin'] = np.sin(2*np.pi*yankees['year']/max(yankees['year']))
yankees['year_cos'] = np.cos(2*np.pi*yankees['year']/max(yankees['year']))
yankees['month_sin'] = np.sin(2*np.pi*yankees['month']/12)
yankees['month_cos'] = np.cos(2*np.pi*yankees['month']/12)
yankees['day_sin'] = np.sin(2*np.pi*yankees['day']/yankees['days_in_month'])
yankees['day_cos'] = np.cos(2*np.pi*yankees['day']/yankees['days_in_month'])
yankees['hour_sin'] = np.sin(2*np.pi*yankees['hour']/24)
yankees['hour_cos'] = np.cos(2*np.pi*yankees['hour']/24)
yankees['minute_sin'] = np.sin(2*np.pi*yankees['minute']/60)
yankees['minute_cos'] = np.cos(2*np.pi*yankees['minute']/60)
yankees['second_sin'] = np.sin(2*np.pi*yankees['second']/60)
yankees['second_cos'] = np.cos(2*np.pi*yankees['second']/60)
```

```python
# Dropping the originial date column and the newly created individual date component columns
yankees = yankees.drop(['DATE_AND_TIME','year', 'month', 'day', 'hour', 'second', 'minute', 'days_in_month'], axis=1)
```

```python
# Assigning values from the 'OFNS_DESC' column to a new column related to sex crimes
sex_values_to_replace = ['SEX CRIMES', 'RAPE', 'PROSTITUTION & RELATED OFFENSES']
sex_new_value = 'Sex Crimes'

for value in sex_values_to_replace:
    yankees['OFNS_DESC'].replace(value, sex_new_value, inplace=True)
```

```python
# Assigning values from the 'OFNS_DESC' column to a new column related to property crimes
property_values_to_replace = ['PETIT LARCENY', 'GRAND LARCENY', 'ROBBERY', 'BURGLARY', 'GRAND LARCENY OF MOTOR VEHICLE',
                            'CRIMINAL TRESPASS','THEFT-FRAUD', 'OFFENSES INVOLVING FRAUD', 'FRAUDS', 'POSSESSION OF STOLEN PROPERTY',
                            'OTHER OFFENSES RELATED TO THEF', 'UNAUTHORIZED USE OF A VEHICLE', 'ARSON', 'THEFT OF SERVICES', 'FRAUDULENT ACCOSTING',
                            'BURGLAR\'S TOOLS', 'PETIT LARCENY OF MOTOR VEHICLE', 'JOSTLING']
property_new_value = 'Property Crimes'

for value in property_values_to_replace:
    yankees['OFNS_DESC'].replace(value, property_new_value, inplace=True)
```

```python
# Assigning values from the 'OFNS_DESC' column to a new column related to violent crimes
violence_values_to_replace = ['ASSAULT 3 & RELATED OFFENSES', 'FELONY ASSAULT', 'HARRASSMENT 2', 'KIDNAPPING & RELATED OFFENSES',
                            'HOMICIDE-NEGLIGENT,UNCLASSIFIE', 'KIDNAPPING']
violence_new_value = 'Violence-Related Crimes'

for value in violence_values_to_replace:
    yankees['OFNS_DESC'].replace(value, violence_new_value, inplace=True)
```

```python
# Assigning values from the 'OFNS_DESC' column to a new column related to public order crimes
public_values_to_replace = ['CRIMINAL MISCHIEF & RELATED OF', 'OFF. AGNST PUB ORD SENSBLTY &', 'OFFENSES AGAINST PUBLIC ADMINI', 'INTOXICATED & IMPAIRED DRIVING', 'ALCOHOLIC BEVERAGE CONTROL LAW',
                            'OFFENSES AGAINST PUBLIC SAFETY', 'DISORDERLY CONDUCT', 'INTOXICATED/IMPAIRED DRIVING']
public_new_value = 'Public Order Crimes'

for value in public_values_to_replace:
    yankees['OFNS_DESC'].replace(value, public_new_value, inplace=True)
```

```python
# Assigning values from the 'OFNS_DESC' column to a new column related to other crimes
other_values_to_replace = ['DANGEROUS DRUGS', 'DANGEROUS WEAPONS', 'MISCELLANEOUS PENAL LAW', 'VEHICLE AND TRAFFIC LAWS', 'FORGERY', 'ADMINISTRATIVE CODE',
                            'OFFENSES AGAINST THE PERSON', 'NYS LAWS-UNCLASSIFIED FELONY', 'OTHER STATE LAWS (NON PENAL LA', 'AGRICULTURE & MRKTS LAW-UNCLASSIFIED',
                            'GAMBLING', 'CHILD ABANDONMENT/NON SUPPORT', 'ENDAN WELFARE INCOMP', 'ESCAPE 3', 'NYS LAWS-UNCLASSIFIED VIOLATION', 'OFFENSES RELATED TO CHILDREN',
                            'OTHER STATE LAWS', 'ABORTION', 'ANTICIPATORY OFFENSES', 'NaN']
other_new_value = 'Other Crimes'

for value in other_values_to_replace:
    yankees['OFNS_DESC'].replace(value, other_new_value, inplace=True)
```

```python
# If there is an alert in range, assign a 1
for i, row in yankees.iterrows():
    # Check if the value in the column is an empty list
    if not row['alerts_in_range']:
        # Assign the value 0
        yankees.at[i, 'alerts_in_range'] = 0
    else:
        # Assign the value 1
        yankees.at[i, 'alerts_in_range'] = 1
```

```python
# Dropping the target variable
X = yankees.drop(columns=['OFNS_DESC'])
X
```

```python
# Assigning dummy variables for categorical features
X = pd.get_dummies(X, columns=['LAW_CAT_CD', 'BORO_NM', 'closest_station'])
```

```python
# Changing the data type for the train line columns
X = X.astype({'alerts_in_range': 'int', 'Line_1': 'int', 'Line_2': 'int', 'Line_4': 'int', 'Line_5': 'int', 'Line_6': 'int',
            'Line_A': 'int', 'Line_B': 'int', 'Line_C': 'int', 'Line_D': 'int', 'Line_G': 'int', 'Line_J': 'int',
            'Line_L': 'int', 'Line_Z': 'int'})
```

```python
# Neccessary imports
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.preprocessing import LabelEncoder
```

```python
# Encoding the y-values

# Create the label encoder
encoder = LabelEncoder()

# Fit the encoder to your dependent variable
encoder.fit(y)

# Transform the dependent variable to numerical values
y_encoded = encoder.transform(y)
```

```python
# These columns had a feature importance of 0 so were dropped
X = X.drop(columns=['year_sin', 'closest_station_Utica Av', 'closest_station_Gates Av','second_sin', 'second_cos',
                    'BORO_NM_BRONX', 'Line_G', 'closest_station_Dyckman St', 'year_cos'])
X = X.drop(columns= ['Line_Z', 'closest_station_Shepherd Av', 'closest_station_Canarsie - Rockaway Parkway', 'closest_station_Classon Av','closest_station_168th St - Washington Heights'])
X = X.drop(columns='Line_J')
X = X.drop(columns=['closest_station_116th St', 'closest_station_135th St', 'closest_station_163rd St - Amsterdam Av'])
X = X.drop(columns='Line_L')
```

```python

seed = 10
X_train, X_test, y_encoded, y_test = train_test_split(X, y_encoded, test_size=0.15, random_state=seed)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# list of class weights
class_weights = [
    {0:1, 1:10, 2:10, 3:10, 4:10},
    {0:1, 1:15, 2:15, 3:15, 4:15},
    {0:1, 1:20, 2:20, 3:20, 4:20},
    {0:1, 1:25, 2:25, 3:25, 4:25},
    {0:1, 1:30, 2:30, 3:30, 4:30},
    {0: 1, 1: 30, 2: 30, 3: 30, 4: 30}, 
    {0: 1, 1: 35, 2: 35, 3: 35, 4: 35},
    {0: 1, 1: 40, 2: 40, 3: 40, 4: 40},
    {0: 1, 1: 45, 2: 45, 3: 45, 4: 45},
    {0: 1, 1: 50, 2: 50, 3: 50, 4: 50}
]
```

```python
# Set the parameters up 
parameters = {
    "n_estimators":[200,300,400,500,600,700],
    "max_depth":[100,150,200,250,300],
    "min_samples_split":[0.00000001, 0.0000001, 0.000001, 0.00001,0.0001,0.01],
    "n_jobs":[-1],
    "max_samples": [0.001, 0.01, 0.1, 1],
    "class_weight": class_weights
}
```

```python
# Initialize the model
rf = RandomForestClassifier()
```

```python
# Initialize grid search 
cv = GridSearchCV(rf,parameters,cv=3)
```

```python
# fit grid search
cv.fit(X_train,y_encoded)
```

```python
# function that returns the best params
def display(results):
    print(f'Best parameters are: {results.best_params_}')
```

```python
# displays the optimal params
display(cv)
```

```python
# Initialize the model with its parameters
# Note: the parameters aren't the same as the optimal parameters listed above. That is because they were eventually calculated manually due to long computation times
rf = RandomForestClassifier(max_depth=675,max_samples=.35,min_samples_split=0.000000001, n_estimators=780, n_jobs=-1, oob_score=True, class_weight={0:1, 1:28, 2:28, 3:28, 4:28})
```

```python
# Fit the model 
rf.fit(X_train, y_encoded)
```

```python
y_pred = rf.predict(X_test)
```

```python
# Gets the accuracy of the model
accuracy_score(y_test, y_pred) 
```

```python
# Create the ROC chart
from sklearn.metrics import roc_auc_score, roc_curve


n_classes = len(y.unique())
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    y_test_binary = np.where(y_test == i, 1, 0)
    y_pred_binary = np.where(y_pred == i, 1, 0)
    fpr[i], tpr[i], _ = roc_curve(y_test_binary, y_pred_binary)
    roc_auc[i] = roc_auc_score(y_test_binary, y_pred_binary)

plt.figure()
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label='Class {0} (AUC = {1:0.2f})'.format(i, roc_auc[i]))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
# Calculate the accuracy on the validation set
val_accuracy = accuracy_score(y_test, y_pred)

print("Validation Accuracy: {:.2f}%".format(val_accuracy * 100))

# Print the OOB score
print("OOB score: ", rf.oob_score_)
```

```python
# Create confusion matrix and classification report 
from sklearn.metrics import confusion_matrix, classification_report

y_pred = rf.predict(X_test)
confusion_mat = confusion_matrix(y_test, y_pred)
print("Confusion Matrix: \n", confusion_mat)
print("\nClassification Report: \n", classification_report(y_test, y_pred))
```

```python
# Prints more accuracy scores 


precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')


print("Precision: {:.2f}".format(precision))
print("Recall: {:.2f}".format(recall))
print("F1-Score: {:.2f}".format(f1))

```
